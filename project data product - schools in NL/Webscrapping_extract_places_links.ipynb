```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import html5lib
import lxml
```

```python
def extract_school_places (url, typesc):
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    components = soup.select('li a')
    places_links = []
    places_names = []
    for component in components: 
        place = component.text
        place_html = place.replace(' ', '%20')
        if typesc == 'basic':
            link = 'https://scholenopdekaart.nl/zoeken/basisscholen?zoektermen='+place_html+'&weergave=Lijst'
        else:
            link = 'https://scholenopdekaart.nl/zoeken/middelbare-scholen?zoektermen='+place_html+'&weergave=Lijst'
        places_links.append(link)
        places_names.append(place)
    return pd.DataFrame({'Name':places_names, 'Link':places_links})
```

```python
basic_sc_links = extract_school_places ('https://scholenopdekaart.nl/basisscholen/', 'basic')
pd.set_option('display.max_colwidth', None)
basic_sc_links.head(20)
```

```python
mb_sc_links = extract_school_places ('https://scholenopdekaart.nl/middelbare-scholen/', 'middle')
pd.set_option('display.max_colwidth', None)
mb_sc_links.head(20)
```

```python
basic_sc_links.to_csv('places_basic_schools.csv')
mb_sc_links.to_csv('places_middle_schools.csv')
```
