```python
import time
from getpass import getpass
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common import NoSuchElementException, ElementClickInterceptedException
import pathlib
import os
from os.path import join
import pandas as pd
import random
import re
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
def percentage_passing_students_middle(dataframe):
    passing_students = []
    errors = []
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

    try:
        for url in dataframe['link']:
            try:
                driver.get('https://scholenopdekaart.nl' + url + 'resultaten/')
                html = driver.page_source
                soup = BeautifulSoup(html)
                charts_elems = soup.find_all(class_= re.compile(r'highcharts-markers highcharts-series-0 highcharts-area-series highcharts-color-0 highcharts-tracker'))
                elems_stud_pas = charts_elems[2]
                
                aria_labels_sp = []
                for element in elems_stud_pas:
                    aria_label = element['aria-label']
                    aria_labels_sp.append(aria_label)
                
                students_who_passed = aria_labels_sp[2].split(' ')[2].replace(',','.').strip('.')

                students_who_passed_dict = {'URL': url, 'Students_who_passed': float(students_who_passed)}
                passing_students.append(students_who_passed_dict)

            except Exception as e:
                error_message = f"Error occurred for URL {url}: {str(e)}"
                print(error_message)
                errors.append({'URL': url, 'Error': error_message})

        passing_students_df = pd.DataFrame(passing_students)
        errors_students_df = pd.DataFrame(errors)
        return passing_students_df, errors_students_df

    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return pd.DataFrame(columns=['URL', 'Percentage_passing_student']), pd.DataFrame(columns=['URL', 'Error'])
```

```python
middle_schools = pd.read_csv('middle_schools.csv')
```

```python
passing_students_df, errors_students_df = percentage_passing_students_middle(middle_schools)
```

```python
passing_students_df.rename(columns={'URL':'link'}, inplace=True)
```

```python
errors_students_df.rename(columns={'URL':'link'}, inplace=True)
```

```python
passing_students_df.shape
```

```python
errors_students_df.shape
```

```python
middle_schools_passing_st = pd.merge(middle_schools, passing_students_df, on='link', how='left')
```

```python
middle_schools_passing_st.to_csv('middle_schools_passing_st.csv', index=False)
```

```python
def get_avg_score_exam_middle(dataframe):
    avg_scores = []
    errors = []
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

    try:
        for url in dataframe['link']:
            try:
                driver.get('https://scholenopdekaart.nl' + url + 'resultaten/')
                html = driver.page_source
                soup = BeautifulSoup(html)
                charts_elems = soup.find_all(class_= re.compile(r'highcharts-markers highcharts-series-0 highcharts-area-series highcharts-color-0 highcharts-tracker'))
                elems_avg_sc = charts_elems[3]
                
                aria_labels_avg_sc = []
                for element in elems_avg_sc:
                    aria_label = element['aria-label']
                    aria_labels_avg_sc.append(aria_label)
                avg_score = aria_labels_avg_sc[2].split(' ')[2].replace(',','.').strip('.')

                avg_score_dict = {'URL': url, 'Average_exam_score': float(avg_score)}
                avg_scores.append(avg_score_dict)

            except Exception as e:
                error_message = f"Error occurred for URL {url}: {str(e)}"
                print(error_message)
                errors.append({'URL': url, 'Error': error_message})

        avg_scores_df = pd.DataFrame(avg_scores)
        errors_scores_df = pd.DataFrame(errors)
        return avg_scores_df, errors_scores_df

    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return pd.DataFrame(columns=['URL', 'Avg_score_exam']), pd.DataFrame(columns=['URL', 'Error'])
```

```python
middle_schools = pd.read_csv('middle_schools.csv')
```

```python
avg_scores_df, errors_scores_df = get_avg_score_exam_middle(middle_schools)
```

```python
avg_scores_df.rename(columns={'URL':'link'}, inplace=True)
errors_scores_df.rename(columns={'URL':'link'}, inplace=True)
```
