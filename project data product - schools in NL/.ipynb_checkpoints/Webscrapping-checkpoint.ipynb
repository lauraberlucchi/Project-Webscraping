```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import html5lib
import lxml
```

```python
def extract_school_places (url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    components = soup.select('li a')
    places_links = []
    for component in components:
        link = component.get('href')
        places_links.append(link)
    return places_links
```

```python
def extract_school_data (place):
    url = 'https://scholenopdekaart.nl'+place
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    components = soup.select('li a')
    places_links = []
    for component in components:
        link = component.get('href')
        places_links.append(link)
    return places_links
```
